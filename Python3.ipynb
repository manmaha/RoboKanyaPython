## 7. Your First Meaningful Python Program : Quadratic Equation Solver

Quadratic Equations are of the type:
ax^2 + bx + c = 0
We solve them by calculating D (the Discriminant)
D = b^2 - 4ac
If D < 0, then the quadratic equation has no real roots
If D > 0, then the quadratic equation has the real roots R1 and R2
where R1 = (-b+sqrt(D))/2a
and R2 = (-b+sqrt(D))/2a

## 8. Loop Structures : For Statement

Sometimes you need to execute a set of statements many times over. This is called Looping or Iteration
Can you think of some reasons why this would be needed:





Python has very powerful iteration and looping structures. We shall study the For statement in this class. The syntax of the For statement is very similar to the If statement

For (Iterable) Condition:
    Block of Code to be Executed
    
Evaluate Condition
If Condition is True, Execute Code, Loop Back to Evaluate Condition
If Condition is False, exit For loop and continue with the Program

Fairly simple right?

Lets see how this works in practise:

Lets input three names and print them out. We can do this in two ways

#### First Method:

```python
first_name=input('Enter First Name ')
second_name=input('Enter Second Name ')
third_name=input('Enter Third Name ')
print('First Name is '+first_name)
print('Second Name is '+second_name)
print('Third Name is '+third_name)
```

Hmmm.... 6 lines of Code - will not work if you have to enter all the names in a class!

Lets try this

#### Second Method


```python
for i in range(1,4):
    name = input('Enter '+str(i)+'st Name')
    print(str(i)+'st Name is '+name)
```

### Isn't this powerful!!
### These 3 lines of codes can input and print any size of data!

There's only one level of confustion - what's this i in range(1,3)??

Turns out that Python has two very powerful constructs called Lists and Iterators - both are very intuitive and easy to learn
