## 5. Introduction to Algorithms and Pesudo Code

An Algorithm is a plan, standarized process or  set of rules to solve a problem. It is specific to a problem, very detailed and covers all possibilities. If you give it to anyone, they should be able to follow it. In fact it is so systematic that machines or computers can use it. Algorithms are hence the basis for all computer programs.


A simple example of an Algorithm:
Taking the average of two numbers
1. Take the First Number
2. Add it to the Second Number and Store the answer in Sum
3. Divide Sum by 2 , this is the Average

Algorithms need to be converted into a program using the right syntax but sometimes writing pseudo code (or a language that looks like a programming language but is not one!) is a good intermediate step.

That will look like this:
1. Input a,b
2. Sum <-- a+b
3. Average = Sum/2
4. Return Average

## 6.1 If Statement in Python






The 'If' statement is a very powerful control statment in Python. 
Syntax : if Condition:
            statements1
         else:
             statements2
Very Easy ----> First the Condition is evaluated: If True then statements1 are executed else statements2 are executed

Questions:
1. What Type of value will the Condition return
2. Can we have a If statement that always executes statements1?

```python
a = 5
b = 6
if a<b :
    print('a less than b')
else :
    print ('a greater than or equal to b')
```

```python
a = 7
b = 6
if a<b :
    print('a less than b')
else :
    print ('a greater than or equal to b')
```

```python
if True:
    print('this statment will always be executed')
else:
    print ('this statement will never be executed')
```

### Python Basics: Indented Code blocks
Python uses indentation to denote a 'block' of code.

The block of code that gets executed after an if statement is an example of an indented code block.

You can use tabs or spaces for indentation. All statements at the same indentation level will be considered as a 'block' of code.
